```python
# !python -m spacy download en_core_web_trf
import re
import spacy
print(spacy.__version__)
```

```python
nlp = spacy.load("en_core_web_trf") # 'en_core_web_trf'
```

```python
# Input paragraph
paragraph = '''
[The target context goes here]
'''

# Remove formatting notations, list item notations, and list numbers/characters
paragraph = re.sub(r'#\s+|##\s+|###\s+|####\s+|\*\*|  \d+\.\s|- |•|◦| \\\(| \\\)', '', paragraph)  # 'str with \n'
paragraph = re.sub(r'\n\n', '. ', paragraph)
paragraph = re.sub(r'\n', ' ', paragraph)
paragraph = re.sub(r'\s+', ' ', paragraph)
paragraph = re.sub(r'\.+', '.', paragraph)

# Process the text with spaCy for sentence splitting
doc = nlp(paragraph) # tokens

# Split the paragraph into sentences
sentences = [sent.text.strip() for sent in doc.sents if sent.text.strip()]  # list of strs
# print(sentences)

# Add numbers to the sentences
sentences = [f"<{i+1}> {s}" for i, s in enumerate(sentences)]

# print
for sent in sentences:
  print(sent)



```
